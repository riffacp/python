```python
hh=np.array([2,3,4,5,6])
uu=np.array([7,8,9,0,2])
kk=np.concatenate((hh,uu))
print(kk)
```

```python
#to change row and column
ww=np.array([[2,3,4,6,2],[5,6,7,7,8]])
dd=np.array([[9,3,2,6,2],[8,6,4,7,8]])
mm=np.stack((ww,dd),axis=1)
print(mm)
```

```python
#combine two array
ww=np.array([[2,3,4,6,2],[5,6,7,7,8]])
dd=np.array([[9,3,2,6,2],[8,6,4,7,8]])
mm=np.stack((ww,dd))
print(mm)
```

```python
ww=np.array([[2,3,4,6,2],[5,6,7,7,8]])
dd=np.array([[9,3,2,6,2],[8,6,4,7,8]])
mm=np.concatenate((ww,dd),axis=1)
print(mm)
```

```python
#row combined comes first
ww=np.array([[2,3,4,6,2],[5,6,7,7,8]])
dd=np.array([[9,3,2,6,2],[8,6,4,7,8]])
mm=np.hstack((ww,dd))
print(mm)
```

```python
#column combined comes first
ww=np.array([[2,3,4,6,2],[5,6,7,7,8]])
dd=np.array([[9,3,2,6,2],[8,6,4,7,8]])
mm=np.vstack((ww,dd))
print(mm)
```
