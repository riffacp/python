```python
from sklearn.preprocessing import LabelEncoder
```

```python
#only string has to covert
l=LabelEncoder()
data['airconditioning']=l.fit_transform(data['airconditioning'])
data['stories']=l.fit_transform(data['stories'])
data['parking']=l.fit_transform(data['parking'])
data['prefarea']=l.fit_transform(data['prefarea'])
data['basement']=l.fit_transform(data['basement'])
data['mainroad']=l.fit_transform(data['mainroad'])
data['hotwaterheating']=l.fit_transform(data['hotwaterheating'])
data['livingroom']=l.fit_transform(data['livingroom'])
data['bedrooms']=l.fit_transform(data['bedrooms'])
data['bathrooms']=l.fit_transform(data['bathrooms'])
data['furnishingstatus']=l.fit_transform(data['furnishingstatus'])

data.head()
```

```python
x=data.drop(['price'],axis=1)
y=data.price
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)

```
